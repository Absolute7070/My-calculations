# Atomic bomb energy scale [1]

The Trinity test in 1945: 

```python
t=0.006  #s 
R=80  #m  
rho= 1.2 #kg/m3 
```

Energy is:

```python
E= R**5 *rho / t**2
```

```python
scientific_notation= '{:e}'.format(E)
print('E =', scientific_notation, 'J')
```

# Relative to energy usage in Finland 2019:

```python
E_f= 1362053e12 #J 
print('Energy relative to yearly energy consumption in Finland:')
print('E/E_f =', E/E_f)
print()
print('Energy relative to monthly energy consumption in Finland:')
print('E/(E_f/12) =', E/(E_f/12))  #relative to energy usage per month
```

# Energy usage per person in Finland: 

```python
E_p= 5.38884e+10 

print('E/E_p =', E/E_p)
```

Which is equivalent to energy consumption of 2026 people a year !!!  

# How much water to be heated from 0 C to 100 C

```python
c= 4.184  #kJ/(kg K)
deltaT= 100 # K 

m = E/(c*deltaT)

print('m =', m)
```

, which is equivalent to 

```python
rho_v= 1000 #kg/m3
print('V =', m/rho_v, 'm^3' )
print('or')
scientific_notation = '{:e}'.format(m/rho_v*10**3)
print('V =', scientific_notation, 'l')
```
